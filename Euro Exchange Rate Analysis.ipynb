# The Exchange Rate between the Euro and U.S Dollar

The dataset shown illustrates the correlation between the Euro and other currencies around the world. For example, a value in the `U.S. Dollar` column might say 1.6228. This means that one Euro is worth 1.6338 dollars that day. The first column has the date for every value. I will use that row to analyze and visualize correlation between currencies across a period of time. 
Before I do any of that, I will have to clean the data to make sure that the numbers are accurate and the data conveys a clear message. I will be using mardowns to illustrate to paint a picture of what I'm trying to accomplish.

The data I will be working with can be found on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020).

### Importing and reading file

```python
#First I'll import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Next I'll go ahead and read the csv file and name the dataframe 'excgange_rates'
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

### Data Cleaning

 I can see that theres a large amount of columns in this dateframe. This is because 40 columns contain currencies I'll be analyzing for this project. I'm going to use the `exchange_rates.info()` to get a better idea of how the columns look in terms of datatypes and string characters.

```python
exchange_rates.info()
```

**DataFrame Break Down**
* This data frame is made up of 5,698 rows and 43 columns. 
* The column names will need to be cleaned.
* The coluumns are filled with different currencies from all over the world. 
* We definitily have null values in the dataframe that need to be addressed. 
* There are 38 columns containing string objects. 
* There are only three columns that contain floats.

**Cleaning the column names**

```python
def clean_cols(columns):
    columns = columns.str.replace('[', '')
    columns = columns.str.replace(']', '')
    columns = columns.str.strip()
    columns = columns.str.replace(' ', '_')
    return columns

exchange_rates.columns = clean_cols(exchange_rates.columns)
    
```

I'll use `exchange_rates.head()` to confirm the column names have been changed. 

```python
exchange_rates.head()
```

It seems like the column names have been cleaned. Therefore, I will proceed with modifying the first column from `Period\Unit` to `Time`. After that, I will go ahead and make it a datetime property. 

```python
exchange_rates.rename(columns={'Period\\Unit:': 'Time'},
                     inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time']) # Changing Time column into a datetime property
exchange_rates.sort_values('Time', inplace=True) # Sorting values according to date
exchange_rates.reset_index(drop=True, inplace=True)
```

I'd like to see the frequency of the correlation of elements between `Time` and `US_dollar` columns.  

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

The dashes have a high frequency of 64. This is most likely because who ever entered the data used the dashes as a place holder for null values. I'll use a boolean to get rid of them. 

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'].astype(float).value_counts()
```

I would like to see the correlation between the Euro and the Dollar represented in graph form. Since there are many values that may interfere with a clear line plot, I plan to group 30 of the values together to determine the mean, which I will use as my y-values.

```python
rolling_mean = euro_to_dollar['US_dollar'].rolling(30).mean()
rolling_mean

plt.plot(euro_to_dollar['Time'], rolling_mean)
plt.show()
```

### We're going to create a visual of the three administrations that occupied the White House from the years 2000 to 2021. 

```python
fig, ax = plt.subplots(figsize=(10,3))

ax.plot(euro_to_dollar['Time'], rolling_mean, alpha=.05)
ax.tick_params(bottom=0, left=0) 
for location in ['right', 'top']:
    ax.spines[location].set_visible(False)
ax.axvline(x=pd.to_datetime('2000-01-04'), color='grey', alpha=0.5)
ax.axvline(x=pd.to_datetime('2008-01-04'), color='grey', alpha=0.5)
ax.axvline(x=pd.to_datetime('2016-01-04'), color='grey', alpha=0.5)
ax.plot(euro_to_dollar['Time'][257:2300], rolling_mean[257:2300], color='#af0b1e', linewidth=2.5)
ax.plot(euro_to_dollar['Time'][2300:4350], rolling_mean[2300:4350], color='#1f77b4', linewidth=2.5)
ax.plot(euro_to_dollar['Time'][4350:5700], rolling_mean[4350:5700], color='#2ca02c', linewidth=2.5)

ax.text(pd.to_datetime('2004-01-04'), 1.7, 'Bush', fontsize=14, color='#af0b1e', horizontalalignment='center')
ax.text(pd.to_datetime('2012-01-04'), 1.7, 'Obama', fontsize=14, color='#1f77b4', horizontalalignment='center')
ax.text(pd.to_datetime('2018-01-04'), 1.7, 'Trump', fontsize=14, color='#2ca02c', horizontalalignment='center')
for y_value in [1.0, 1.2, 1.4, 1.6]:
    ax.hlines(y_value, xmin=euro_to_dollar['Time'].iloc[0], xmax=euro_to_dollar['Time'].iloc[-1], color='grey', alpha=0.06)
plt.show()
```

We can see the 2008 crash peak in the beginning of Obama's term. It started trending down as the years went on. The housing market crash took a huge toll on the correlation between the dollar and the euro.

#### Now I'm going to compare all three currencies using a grid chart of line graphs.

```python
Chinese_rolling_mean = euro_to_yuan['Chinese_yuan_renminbi'].rolling(30).mean()
Jap_rolling_mean = euro_to_yen['Japanese_yen'].rolling(30).mean()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(12, 14))


ax1.plot(euro_to_yuan['Time'], Chinese_rolling_mean)
ax1.text(y=12, x=pd.to_datetime('2010-01-04'), s='China',fontsize=16, color='#af0b1e')
ax1.axvline(x=pd.to_datetime('2008-07-04'), color='grey', alpha=0.5)

ax2.plot(euro_to_yen['Time'], Jap_rolling_mean)
ax2.text(y=175, x=pd.to_datetime('2010-01-04'), s='Japan',fontsize=16, color='#1f77b4')
ax2.axvline(x=pd.to_datetime('2008-07-04'), color='grey', alpha=0.5)

ax3.plot(euro_to_dollar['Time'], rolling_mean)
ax3.text(y=1.65, x=pd.to_datetime('2010-01-04'), s='USA',fontsize=16, color='#2ca02c')
ax3.axvline(x=pd.to_datetime('2008-07-04'), color='grey', alpha=0.5)
plt.show()
```

These graphs paint a picture about the world economy and how one event can affect the whole thing. For example the 2008 crash can be seen in all three countries. One can argue the country that caused the crash suffered the most to get it's economy back on track. That country was the U.S.A..  The other two put in massive fiscal stimulus packages that seemed to help. On the other hand, the U.S. Dollar looked like it's making a gradual recovery from the 2008 crash.
